##1.4. Dashboard tổng quan: Doanh thu trung bình theo ngày trong tháng

Doanh thu trung bình hàng ngày dao động khoảng từ 10M đến gần 14M VND.Có sự biến động theo từng ngày, có ngày doanh thu đột biến cao hơn rõ rệt.Có thể xem xét các yếu tố ngoại cảnh hoặc chương trình khuyến mãi để giải thích biến động này.

```python
# Đảm bảo cột datetime
df['Thời gian tạo đơn'] = pd.to_datetime(df['Thời gian tạo đơn'])

# Tạo cột 'Tháng' và 'Ngày trong tháng'
df['Tháng'] = df['Thời gian tạo đơn'].dt.month
df['Ngày trong tháng'] = df['Thời gian tạo đơn'].dt.day

# Bước 1: Tính tổng doanh thu và lượng bán theo từng (Tháng, Ngày trong tháng)
monthly_daily_total = df.groupby(['Tháng', 'Ngày trong tháng']).agg({
    'Thành tiền': 'sum',
    'SL': 'sum'
}).reset_index()

# Bước 2: Tính trung bình doanh thu và lượng bán cho từng 'Ngày trong tháng' qua các tháng
avg_by_day = monthly_daily_total.groupby('Ngày trong tháng').agg({
    'Thành tiền': 'mean',
    'SL': 'mean'
}).reset_index().rename(columns={
    'Thành tiền': 'Doanh thu trung bình',
    'SL': 'Lượng bán trung bình'
})

# Chuyển doanh thu trung bình sang hàng triệu và làm tròn 1 chữ số thập phân
avg_by_day['Doanh thu trung bình (triệu)'] = (avg_by_day['Doanh thu trung bình'] / 1_000_000).round(1)

# Làm tròn lượng bán trung bình tới số nguyên
avg_by_day['Lượng bán trung bình (đơn vị)'] = avg_by_day['Lượng bán trung bình'].round(0).astype(int)

# Chọn cột để hiển thị
result = avg_by_day[['Ngày trong tháng', 'Doanh thu trung bình (triệu)', 'Lượng bán trung bình (đơn vị)']]

result
```

##1.5. Dashboard tổng quan: Xác suất bán hàng theo nhóm hàng

Nhóm hàng có xác suất bán hàng cao nhất là Trà mix (54.5%), tiếp theo là Trà hoa (54.4%), Trà củ, quả sấy (53.3%).Mặc dù Trà mix có xác suất bán hàng cao nhưng doanh thu lại thấp hơn Trà hoa và Trà củ, quả sấy.Đây có thể là tín hiệu cho thấy trà mix được bán phổ biến hơn, nhưng giá trị đơn hàng thấp hơn hoặc bán với giá thấp hơn.

```python
# Tính số đơn hàng duy nhất theo nhóm hàng
don_hang_theo_nhom = df.groupby('Mã nhóm hàng')['Mã đơn hàng'].nunique()

# Tổng số đơn hàng duy nhất toàn bộ
tong_don_hang = df['Mã đơn hàng'].nunique()

# Tính xác suất bán hàng theo nhóm hàng nhân 100 (phần trăm), làm tròn 3 chữ số thập phân
xac_suat_ban_hang = ((don_hang_theo_nhom / tong_don_hang) * 100).round(1)

# Tạo DataFrame kết quả
df_xac_suat = xac_suat_ban_hang.reset_index().rename(columns={
    'Mã nhóm hàng': 'Nhóm hàng',
    'Mã đơn hàng': 'Xác suất bán hàng (%)'
})

df_xac_suat
```

##1.6. Dashboard tổng quan: Lượng bán trung bình theo từng ngày trong tháng

Lượng bán hàng và số lượng khách hàng khá ổn định trong suốt tháng.Tuy nhiên có sự sụt giảm đáng chú ý vào cuối tháng (ngày 30-31), có thể do yếu tố cuối tháng hoặc ngày nghỉ lễ.Nên phân tích thêm để hiểu nguyên nhân và có phương án thúc đẩy bán hàng cuối kỳ.


```python
# Đảm bảo cột 'Thời gian tạo đơn' là datetime
df['Thời gian tạo đơn'] = pd.to_datetime(df['Thời gian tạo đơn'])

# Tạo cột 'Ngày trong tháng'
df['Ngày trong tháng'] = df['Thời gian tạo đơn'].dt.day

# Tính tổng lượng bán (SL) và số đơn hàng (Mã đơn hàng) duy nhất theo từng ngày thực tế (date)
daily_agg = df.groupby(df['Thời gian tạo đơn'].dt.date).agg({
    'SL': 'sum',
    'Mã đơn hàng': 'nunique'  # số lượng đơn hàng duy nhất mỗi ngày
}).reset_index().rename(columns={'Thời gian tạo đơn': 'Ngày'})

# Tính trung bình lượng bán theo ngày trong tháng
avg_luong_ban = daily_agg.groupby(daily_agg['Ngày'].apply(lambda x: x.day)).agg({
    'SL': 'mean'
}).rename(columns={'SL': 'Lượng bán trung bình'}).reset_index()

# Tính tổng số đơn hàng (khách hàng) theo ngày trong tháng
tong_khach_hang = daily_agg.groupby(daily_agg['Ngày'].apply(lambda x: x.day)).agg({
    'Mã đơn hàng': 'sum'
}).rename(columns={'Mã đơn hàng': 'Số lượng khách hàng'}).reset_index()

# Gộp 2 bảng lại theo 'Ngày trong tháng'
result = pd.merge(avg_luong_ban, tong_khach_hang, left_on='Ngày', right_on='Ngày')

# Đổi tên cột 'Ngày' thành 'Ngày trong tháng'
result = result.rename(columns={'Ngày': 'Ngày trong tháng'})

result
```

```python
import matplotlib.pyplot as plt
from IPython.display import display

# Đọc dữ liệu
df = pd.read_csv('/content/drive/MyDrive/File/data_ggsheet.csv')

# Tính tổng doanh thu theo Nhóm hàng
group_rev = df.groupby('Mã nhóm hàng')['Thành tiền'].sum().reset_index()
group_rev = group_rev.sort_values(by='Thành tiền', ascending=False)

# Thêm cột % và % cộng dồn
group_rev['Tỷ lệ (%)'] = (group_rev['Thành tiền'] / group_rev['Thành tiền'].sum()) * 100
group_rev['% cộng dồn'] = group_rev['Tỷ lệ (%)'].cumsum()

# Hiển thị bảng Pareto theo Nhóm hàng
print("Bảng Pareto doanh thu theo nhóm hàng:")
display(group_rev)

# Tính tổng doanh thu theo Mặt hàng
item_rev = df.groupby('Tên mặt hàng')['Thành tiền'].sum().reset_index()
item_rev = item_rev.sort_values(by='Thành tiền', ascending=False)

# Thêm cột % và % cộng dồn
item_rev['Tỷ lệ (%)'] = (item_rev['Thành tiền'] / item_rev['Thành tiền'].sum()) * 100
item_rev['% cộng dồn'] = item_rev['Tỷ lệ (%)'].cumsum()

# Hiển thị bảng Pareto theo Mặt hàng
print("Bảng Pareto doanh thu theo mặt hàng:")
display(item_rev)
```

Về mặt khai thác thực tế, có thể tập trung xây dựng nhóm combo ưu tiên dựa trên các cặp có lượt mua từ 1000 trở lên như “Bột cần tây & Trà gừng” hay “Cam lát & Trà gừng”. ##2.8. Dashboard phân tích giỏ hàng: Biến động mua kèm theo tháng của các combo sản phẩm.

Biểu đồ trên cho thấy rõ ràng xu hướng tăng mạnh số lượt mua kèm đối với các combo sản phẩm trong những tháng cuối năm, đặc biệt từ tháng 8 đến tháng 11. Nổi bật nhất là combo “Bột cần tây & Trà gừng”, với số lượt mua tăng liên tục và đạt đỉnh hơn 900 lượt trong tháng 11, sau đó có xu hướng giảm nhẹ trong tháng 12. Điều này có thể phản ánh ảnh hưởng của yếu tố mùa vụ — thời tiết lạnh hơn khiến khách hàng có xu hướng chọn các sản phẩm tốt cho sức khỏe, thanh lọc và tăng đề kháng. Các combo như “Cam lát & Trà gừng”, “Trà cam sả quế & Trà gừng” hay “Bột cần tây & Cam lát” cũng ghi nhận mức tăng đáng kể trong cùng giai đoạn.

Dữ liệu cũng cho thấy các combo có yếu tố "thanh nhiệt – hỗ trợ tiêu hóa" hoặc "giải độc – tăng đề kháng" có khả năng được ưu tiên hơn trong quyết định mua kèm. Điều này mở ra cơ hội định hướng chiến lược marketing tập trung vào mùa thu và đông, đồng thời có thể tối ưu gợi ý combo trên nền tảng bán hàng.

```python
# Chuẩn hóa định dạng thời gian thành tháng
df['Thời gian tạo đơn'] = pd.to_datetime(df['Thời gian tạo đơn'], errors='coerce')
df['Tháng'] = df['Thời gian tạo đơn'].dt.strftime('%m')

# Gom mặt hàng theo từng đơn hàng + tháng
grouped = df.groupby(['Mã đơn hàng', 'Tháng'])['Tên mặt hàng'].apply(list).reset_index()

# Đếm các cặp sản phẩm theo từng tháng
monthly_pair_counter = []
for _, row in grouped.iterrows():
    products = sorted(set(row['Tên mặt hàng']))
    month = row['Tháng']
    for pair in combinations(products, 2):
        monthly_pair_counter.append((pair[0], pair[1], month))

# Tạo DataFrame tần suất theo tháng
df_monthly_pairs = pd.DataFrame(monthly_pair_counter, columns=['Sản phẩm A', 'Sản phẩm B', 'Tháng'])
df_monthly_pairs['Số lần mua kèm'] = 1

# Tổng hợp số lần mua kèm theo cặp sản phẩm và tháng
pair_by_month = df_monthly_pairs.groupby(['Sản phẩm A', 'Sản phẩm B', 'Tháng'])['Số lần mua kèm'].sum().reset_index()
display(pair_by_month)
#pair_by_month.to_csv('/content/drive/MyDrive/File/pair_by_month.csv', index=False, encoding='utf-8-sig')
```

##3.1. Dashboard phân tích khách hàng: Row card thể hiện chỉ số tổng quan.

Về phần khách hàng, nhằm mong muốn phân tích xem được hành vi của khách hàng hay nói đúng hơn là khả năng quay lại của khách hàng, thì nhóm thực hiện đi sâu vào phân tích các khía cạnh xung quanh khách hàng quay lại mua sản phẩm. Đầu tiên với dữ liệu ta có được 26.575 khách hàng quay lại trong năm 2022 với tỷ lệ quay lại là 49,64% Đem lại 4071 triệu VND cho doanh thu của cửa hàng


```python
orders_per_customer = df.groupby('Mã khách hàng')['Mã đơn hàng'].nunique().reset_index()
orders_per_customer.columns = ['Mã khách hàng', 'Số đơn hàng']
orders_per_customer['Phân loại'] = orders_per_customer['Số đơn hàng'].apply(
    lambda x: 'Khách quay lại' if x > 1 else 'Khách mới'
)

# Tính tỷ lệ khách hàng quay lại
returning_ratio = orders_per_customer['Phân loại'].value_counts(normalize=True) * 100
ty_le_quay_lai = returning_ratio.get('Khách quay lại', 0)

# Lọc khách quay lại
returning_customers = orders_per_customer[orders_per_customer['Phân loại'] == 'Khách quay lại']['Mã khách hàng']
df_returning = df[df['Mã khách hàng'].isin(returning_customers)]

# Tính doanh thu và số đơn hàng của khách quay lại
doanh_thu = df_returning['Thành tiền'].sum()
so_don = df_returning['Mã đơn hàng'].nunique()

# In ra kết quả
print("Tỷ lệ khách hàng quay lại:", round(ty_le_quay_lai, 2), "%")
print("Tổng doanh thu từ khách quay lại:", round(doanh_thu, 0))
print("Tổng số đơn hàng từ khách quay lại:", so_don)
```

##3.2. Dashboard phân tích khách hàng: Pareto chart thể hiện số lượng khách hàng quay lại qua các tháng.

Tập trung vào các tháng cao điểm từ 8 đến 12 đóng góp khoảng 63,57% doanh thu. Ngoài ra thì lượng khách hàng quay lại cũng có xu hướng tăng dần qua từng tháng

Nhìn vào biến động của khách hàng quay lại theo tháng, ta có thể thấy được xu hướng của khách hàng khi họ bắt đầu quay lại mua hàng nhiều hơn ở các tháng cuối năm, bắt đầu từ tháng 7 trở đi thì lượng khách hàng tăng vọt từ 1300 khách hàng lên đến trên 2000 khách hàng và vẫn duy trì xu hướng tăng dần theo thời gian cho đến cuối tháng

```python
# Chuyển cột 'Thời gian tạo đơn' thành định dạng datetime
df['Thời gian tạo đơn'] = pd.to_datetime(df['Thời gian tạo đơn'])

# Trích xuất tháng từ cột 'Thời gian tạo đơn'
df['Tháng'] = df['Thời gian tạo đơn'].dt.month

# Tìm khách hàng quay lại (có nhiều hơn một đơn hàng)
customer_orders = df.groupby('Mã khách hàng')['Mã đơn hàng'].nunique()
returning_customers = customer_orders[customer_orders > 1].index

# Lọc dữ liệu chỉ giữ các đơn hàng từ khách hàng quay lại
df_returning = df[df['Mã khách hàng'].isin(returning_customers)]

# Đếm số lượng khách hàng quay lại duy nhất theo từng tháng
monthly_returning_customers = df_returning.groupby('Tháng')['Mã khách hàng'].nunique().reset_index()

# Tính tổng doanh thu của khách hàng quay lại theo từng tháng
monthly_revenue = df_returning.groupby('Tháng')['Thành tiền'].sum().reset_index()

# Gộp dữ liệu số lượng khách hàng quay lại và doanh thu
monthly_data = monthly_returning_customers.merge(monthly_revenue, on='Tháng', how='left')

# Đặt lại tên cột
monthly_data.columns = ['Tháng', 'Số lượng khách hàng quay lại', 'Doanh thu']

# Đảm bảo tất cả các tháng từ 1 đến 12 được hiển thị (điền 0 nếu không có)
all_months = pd.DataFrame({'Tháng': range(1, 13)})
monthly_data = all_months.merge(monthly_data, on='Tháng', how='left').fillna(0)

# Tính phần trăm doanh thu
total_revenue = monthly_data['Doanh thu'].sum()
monthly_data['Phần trăm doanh thu'] = (monthly_data['Doanh thu'] / total_revenue * 100).round(2)

# Chuyển cột 'Tháng' thành định dạng "Tháng 01", "Tháng 02",...
monthly_data['Tháng'] = monthly_data['Tháng'].apply(lambda x: f'Tháng {x:02d}')

# Chuyển cột 'Số lượng khách hàng quay lại' và 'Doanh thu' thành kiểu số nguyên
monthly_data['Số lượng khách hàng quay lại'] = monthly_data['Số lượng khách hàng quay lại'].astype(int)
monthly_data['Doanh thu'] = monthly_data['Doanh thu'].astype(int)

# Hiển thị DataFrame
print(monthly_data)
```
